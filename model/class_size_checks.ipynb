```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
import os
import boto3
```

```python
def get_bin_plot(meta_data, y_max):
    """Take a labeled meta dataset and y axis value and 
    plot histograms of bins by country."""
    
    fig = plt.figure(figsize=(12,6))

    country_list = [c for c in meta_data["countries"].unique()]

    for country, num in zip(country_list, range(1,6)):
        ax = fig.add_subplot(2,3,num)
        ax.hist(meta_data["label"][meta_data["countries"] == country])
        ax.set_title(country)
        ax.set_ylim([0, y_max])
        ax.set_xlim([0, 4])
        ax.set_xticks([0,1,2,3,4]) 

    plt.tight_layout()
    plt.show()
```
