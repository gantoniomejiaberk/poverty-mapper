```python
import os
from glob import iglob
import pandas as pd
import numpy as np
from sklearn.utils import resample
```

```python
train_path = './**/*_train.csv'
```

```python
train_file_list = [f for f in iglob(train_path, recursive=True) if os.path.isfile(f)]
```

```python
for train_file in train_file_list:
    base_file = pd.read_csv(train_file)
    value_counts = base_file.label.value_counts()
    max_value = np.max(value_counts)
    made_changes = False
    print("Processing ", train_file )
    #print("Before: \n", value_counts)
    for label, class_to_resample in value_counts[1:].items():
        n_samples = max_value - class_to_resample
        if n_samples > 0:
            made_changes = True
            data_for_label = base_file[base_file.label == label]
            additional_samples = resample(data_for_label, 
                                     replace=True,
                                     n_samples=n_samples,
                                     random_state=111)
            base_file = pd.concat([base_file, additional_samples])
    value_counts = base_file.label.value_counts()  
    if made_changes:
        base_file.to_csv(train_file, index=False)
        print("After \n", value_counts)
    else:
        print("File unchanged ", train_file)
    
```
