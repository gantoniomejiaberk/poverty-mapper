```python
import urllib.request as urllib2
import json
import pandas as pd
import geopandas as gpd
import pycountry 
import matplotlib as plt
import mapclassify
import folium
```

```python
# Load country shape files
country_shapes = gpd.read_file("s3://w210-poverty-mapper/shape_files/raw_data/ipums_world_shapes/world_countries_2020.shp")
```

```python
# Keep if country in list
country_list = ["Bangladesh", "Nepal", "Tajikistan", "Timor Leste", "Philippines"]

country_shapes_subset = country_shapes[country_shapes["CNTRY_NAME"].isin(country_list)]
country_shapes_subset
```

```python
# Get country map
m = folium.Map([24, 90], zoom_start=7)


for _, r in country_shapes_subset.iterrows():
    sim_geo = gpd.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                           style_function=lambda x: {"fillColor": "blue", "lineColor": "blue"})
    folium.Popup(r['CNTRY_NAME']).add_to(geo_j)
    geo_j.add_to(m)
m
```
